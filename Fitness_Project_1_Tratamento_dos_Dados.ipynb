## 2. Separação e Tratamento

Para facilitar o tratamento dos dados, a metodologia escolhida será de particionar a base de dados de acordo com a fonte de cada entrada, tornando mais fácil a manipulação do dado.

Como os lançamentos são feitos individualmente por cada aplicativo, o que faremos será: agrupar por aplicativo, tratar os dataframes separadamente, pivotar as linhas em colunas e finalmente reagrupá-lo usando a data como agregador em um novo DataFrame.

### 2.1. My Fitness Pal (Food Tracker)

Como esse é o aplicativo de "food tracking", vamos excluir as colunas de atividade física. Além disso, a granularidade dos dados que queremos é diária, assim, vamos excluir as informações de hora da coluna 'creationDate'.

```python
myfitnesspal = df.loc[df['sourceName'] == 'MyFitnessPal']
```

```python
myfitnesspal.drop(columns=['creationDate', 'endDate', 'duration','sourceName', 'durationUnit', 'dateComponents','activeEnergyBurned',\
                           'activeEnergyBurnedGoal','activeEnergyBurnedGoal','workoutActivityType', 'activeEnergyBurnedUnit'], inplace=True)
myfitnesspal.startDate = pd.to_datetime(myfitnesspal['startDate']).dt.strftime('%Y-%m-%d')

myfitnesspal.head()
```

```python
myfitnesspal = myfitnesspal.rename(columns={
    'value': 'valor',
    'type': 'tipo',
    'startDate': 'data',
    'unit': 'unidade' 
})
myfitnesspal.head()
```

Descobrindo os tipos de lançamentos que o My Fitness Pal faz:

```python
myfitnesspal['tipo'].value_counts()
```

Para efeitos de análise, vamos ficar com somente quatro elementos: Calorias, Proteínas, Carboidratos e Gorduras Totais.

Vamos substituir os nomes das entradas com nomes mais simples, traduzidos e com as devidas unidades especificadas.

Assim, podemos excluir a coluna 'unidade'.

```python
myfitnesspal = myfitnesspal[myfitnesspal['tipo'].isin(['HKQuantityTypeIdentifierDietaryEnergyConsumed', 'HKQuantityTypeIdentifierDietaryFatTotal',\
                                                       'HKQuantityTypeIdentifierDietaryCarbohydrates', 'HKQuantityTypeIdentifierDietaryProtein'])]
myfitnesspal.tipo.value_counts()                                                      
```

```python
myfitnesspal.tipo = myfitnesspal.tipo.map({
    'HKQuantityTypeIdentifierDietaryEnergyConsumed': 'Calorias Consumidas (kcal)',
    'HKQuantityTypeIdentifierDietaryFatTotal': 'Gorduras (g)',
    'HKQuantityTypeIdentifierDietaryCarbohydrates': 'Carboidratos (g)',
    'HKQuantityTypeIdentifierDietaryProtein': 'Proteinas (g)' 
})

myfitnesspal.drop(columns='unidade', inplace=True)

myfitnesspal.head()
```

Para que possamos agregar os valores diários, vamos tratar a coluna 'valor' para que eles possam ser somados.

```python
myfitnesspal['valor'] = pd.to_numeric(myfitnesspal['valor']).round(0)
myfitnesspal = myfitnesspal.groupby(['data','tipo']).agg({'valor': 'sum'})
myfitnesspal.tail(4)
```

Com o dataset agregado, vamos pivotar para deixa-lo pronto para a futura reunião.

```python
myfitnesspal = myfitnesspal.reset_index()
myfitnesspal = myfitnesspal.pivot(index='data',columns= 'tipo', values='valor')
myfitnesspal = myfitnesspal.reset_index()
myfitnesspal.tail(4)
```

###2.3. Fitdays (Balança de Bioimpedância)

Como padrão, desde o início do projeto meu peso é auferido após acordar e fazer as necessidades. Assim, a variável "tempo" não é necessária, assim como outras colunas do DataFrame. 

```python
fitdays = df.loc[df['sourceName'] == 'Fitdays']
```

```python
fitdays.head(2)
```

```python
fitdays.drop(columns=['startDate', 'endDate', 'duration','unit', 'sourceName', 'durationUnit',\
                        'workoutActivityType', 'dateComponents','activeEnergyBurned',\
                        'activeEnergyBurnedGoal','activeEnergyBurnedUnit'], inplace=True)
fitdays.creationDate = pd.to_datetime(fitdays['creationDate']).dt.strftime('%Y-%m-%d')

fitdays.head(2)
```

```python
fitdays = fitdays.rename(columns={
    'value': 'valor',
    'type': 'tipo',
    'creationDate': 'data'
})
fitdays.head(2)
```

```python
fitdays.tipo.value_counts()
```

Vamos excluir algumas entradas irrelevantes como IMC (ou BMI, em inglês) e frequência cardíaca.

```python
fitdays = fitdays[fitdays['tipo'].isin(['HKQuantityTypeIdentifierBodyMass', 'HKQuantityTypeIdentifierBodyFatPercentage',\
                                             'HKQuantityTypeIdentifierLeanBodyMass'])]
fitdays.tipo.value_counts()   
```

Renomeando as entradas

```python
fitdays.tipo = fitdays.tipo.map({
    'HKQuantityTypeIdentifierBodyMass': 'Peso (kg)',
    'HKQuantityTypeIdentifierBodyFatPercentage': '%Gordura',
    'HKQuantityTypeIdentifierLeanBodyMass': 'Massa Magra (kg)'
})

fitdays.head(10)
```

Como temos algumas entradas repetidas vamos ficar somente com as primeiras medições

```python
fitdays.drop_duplicates(subset=["data", "tipo"], keep='last', inplace=True)
fitdays.head()
```

Com os tratamentos feitos, passemos para a parte de agregação e pivoteamento

```python
fitdays['data'] = pd.to_datetime(fitdays['data'])
fitdays['valor'] = pd.to_numeric(fitdays['valor'])
```

```python
fitdays.set_index('data')
```

```python
fitdays.groupby(['data','tipo']).agg({'valor':'sum'})
```

```python
fitdays = fitdays.reset_index()
fitdays = fitdays.pivot(index='data',columns= 'tipo', values='valor')
fitdays = fitdays.reset_index()
fitdays.tail(4)
```

#### 2.3.1. Ajuste nas colunas



Ajustando a coluna "%Gordura": além da multiplicação por 100, vamos corrigir uma falha do equipamento. Isso acontece pois ele so possui 4 eletrodos, ambos nos pés. Dessa forma, a corrente elétrica responsável pela medição acaba passando somente pelas pernas, dificultando uma medição mais precisa. 

Após uma avaliação visual e com ajuda de uma tabela de referência, vamos, arbitráriamente, reduzir uniformemente o valor de '%Gordura" para todas as entradas em 8%. 

No que se refere a peso, em comparação com outras balanças, o modelo atual não difere muito. Assim, não haverá nenhum ajuste na coluna "Peso (kg)".

Na coluna "Massa Magra (kg)" haverá somente um arredondamento.

```python
fitdays['%Gordura'] = fitdays['%Gordura'] * 100 - 8
fitdays['Massa Magra (kg)'] = fitdays['Massa Magra (kg)'].round(1)
```

```python
fitdays.head()
```

Analisando a primeira linha, os dados de '%gordura' e 'Massa Magra' estão discrepantes com o restante das entradas. Isso foi causado por uma falha na leitura por parte da balança por ser a primeira vez na qual estava me pesando.

```python
fitdays.drop(index=0, axis=1, inplace=True)
fitdays = fitdays.reset_index()
fitdays.drop(columns='index', axis=0, inplace=True)
fitdays.head()
```

## 3. Concatenação dos DataFrames

Antes que a concatenação possa ser feita, vamos confirmar se todas as colunas 'data' dos dataframes estão em formato 'datetime'

```python
myfitnesspal.info()
mifitness.info()
fitdays.info()
zepplife.info()
nike.info()
controle.info()
```

```python
dataframes = [zepplife, fitdays, controle, myfitnesspal, mifitness, nike]
```

```python
for dataframe in dataframes:
  dataframe = dataframe.sort_values(by='data')
  dataframe['data'] = pd.to_datetime(dataframe['data'])
```

```python
dataset = dataframes[0]
for i in range(1, len(dataframes)):
  dataset = pd.merge(dataset, dataframes[i], on='data', how='outer')

dataset.sort_values(by='data', ascending=True)
```

```python
columns = ['Peso (kg)', '%Gordura', 'Massa Magra (kg)']
```

```python
columns = ['Peso (kg)', '%Gordura', 'Massa Magra (kg)']

for col in columns:
  dataset[f"{col}_x"].fillna(dataset[f"{col}_y"], inplace=True)
```

```python
dataset.head()
```

```python
dataset['Peso (kg)_x'].fillna(dataset['Peso (kg)'], inplace=True)
```

```python
dataset['Peso (kg)_x']
```

```python
dataset.drop(columns=['Peso (kg)_y', 'Massa Magra (kg)_y','Peso (kg)_y','Peso (kg)'], inplace=True)
dataset = dataset.rename(columns={'Peso (kg)_x': 'Peso (kg)'})
```

```python
dataset = dataset.reset_index()
dataset.drop(columns='index', inplace=True)
```

```python
dataset = dataset.sort_values(by='data', ascending=True)
```
